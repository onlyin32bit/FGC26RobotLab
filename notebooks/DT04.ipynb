# DT04 — Tăng tốc từ 0 lên 75%

Câu hỏi: drivetrain đạt 75% tốc độ nhanh và cân bằng đến mức nào? Đo rise time 10–90%, sai lệch hai bên, heading đỉnh, dòng điện đỉnh và sụt áp.

```python
from pathlib import Path
import sys

root = Path.cwd().resolve()
if root.name == 'notebooks':
    root = root.parent
sys.path.insert(0, str(root / 'src'))

import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Markdown, display
from analysis.io import load_runs
from analysis.metrics import rise_time_10_90, summarize_runs
from analysis.notebook import show_full_data
from analysis.plots import plot_heading, plot_power, plot_velocity, set_plot_style
from analysis.validation import validate_data

set_plot_style()
test_id = 'DT04'
```

```python
data = load_runs(root / 'data' / 'raw' / test_id)
if data.empty:
    display(Markdown('**Chưa có dữ liệu DT04.**'))
else:
    display(validate_data(data))
    display(summarize_runs(data))
    response = []
    for run_id, run in data.groupby('run_id', sort=False):
        speed = (run.left_actual_tps.abs() + run.right_actual_tps.abs()) / 2
        response.append({'run_id': run_id, 'rise_time_10_90_s': rise_time_10_90(run.time_s, speed)})
        plot_velocity(run, f'{run_id} — vận tốc'); plt.show()
        plot_heading(run, f'{run_id} — heading'); plt.show()
        plot_power(run, f'{run_id} — điện'); plt.show()
    display(pd.DataFrame(response))
    show_full_data(data)
```

## Kết luận

Căn thời gian tại `COMMAND_START` nếu marker có mặt. So sánh rise time, dòng đỉnh và sụt áp giữa các configuration; không đánh đổi an toàn lấy một run nhanh nhất.
