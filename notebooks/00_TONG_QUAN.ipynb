# Tổng quan chương trình thử nghiệm Robot 01

Notebook này tổng hợp bảng `data/run_summary.csv`. Nó không thay thế notebook
chi tiết của từng kịch bản.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from IPython.display import Markdown, display

GOC = Path.cwd().resolve()
if GOC.name == "notebooks":
    GOC = GOC.parent
sys.path.insert(0, str(GOC / "src"))

from fgc_analysis.io import tai_tom_tat

tom_tat = tai_tom_tat(GOC / "data" / "run_summary.csv")
display(tom_tat)
```

## Tiến độ theo kịch bản

```python
if tom_tat.empty:
    display(Markdown("**Chưa có hàng kết quả trong `data/run_summary.csv`.**"))
else:
    tien_do = (
        tom_tat.groupby("test_id", dropna=False)
        .agg(
            so_lan_chay=("run_id", "count"),
            so_lan_dat=("pass", lambda cot: cot.astype(str).str.upper().eq("PASS").sum()),
        )
        .reset_index()
    )
    tien_do["ty_le_dat"] = tien_do["so_lan_dat"] / tien_do["so_lan_chay"]
    display(tien_do)

    hinh, truc = plt.subplots(figsize=(10, 5))
    sns.barplot(data=tien_do, x="test_id", y="ty_le_dat", ax=truc)
    truc.set(
        title="Tỷ lệ đạt theo kịch bản",
        xlabel="Mã thử nghiệm",
        ylabel="Tỷ lệ đạt",
        ylim=(0, 1),
    )
    hinh.tight_layout()
    plt.show()
```
