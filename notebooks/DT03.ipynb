# DT03 — Chạy chậm và độ trôi

Câu hỏi: ngưỡng tốc độ thấp nhất robot có thể đi thẳng ổn định là bao nhiêu? Mục tiêu là xác định ngưỡng di chuyển tin cậy và độ lệch ngang ≤ 50 mm/m.

```python
from pathlib import Path
import sys

root = Path.cwd().resolve()
if root.name == 'notebooks':
    root = root.parent
sys.path.insert(0, str(root / 'src'))

import matplotlib.pyplot as plt
from IPython.display import Markdown, display
from analysis.io import load_runs
from analysis.metrics import summarize_runs
from analysis.notebook import show_full_data
from analysis.plots import plot_heading, plot_pid, plot_power, plot_velocity, set_plot_style
from analysis.validation import validate_data

set_plot_style()
test_id = 'DT03'
```

```python
data = load_runs(root / 'data' / 'raw' / test_id)
if data.empty:
    display(Markdown('**Chưa có dữ liệu DT03.**'))
else:
    display(validate_data(data))
    display(summarize_runs(data))
    for run_id, run in data.groupby('run_id', sort=False):
        plot_velocity(run, f'{run_id} — vận tốc'); plt.show()
        plot_heading(run, f'{run_id} — heading'); plt.show()
        plot_power(run, f'{run_id} — điện'); plt.show()
        if {'heading_p', 'heading_i', 'heading_d', 'heading_total'}.issubset(run):
            plot_pid(run, f'{run_id} — PID heading'); plt.show()
    show_full_data(data)
```

So sánh theo lệnh 0,05–0,25. Nêu ngưỡng tốc độ bắt đầu mất ổn định, biên độ hiệu chỉnh, chênh vận tốc trái–phải, dòng điện và các run bị loại.
