# DT02 — Chạy thẳng 2 mét

Câu hỏi: robot giữ hướng và đường tâm tốt đến mức nào ở các tốc độ? Mục tiêu là sai số quãng đường ≤ 3%, RMSE heading ≤ 1,5° và độ lệch ngang trong giới hạn nhóm đã chọn.

```python
from pathlib import Path
import sys

root = Path.cwd().resolve()
if root.name == 'notebooks':
    root = root.parent
sys.path.insert(0, str(root / 'src'))

import matplotlib.pyplot as plt
from IPython.display import Markdown, display
from analysis.io import load_runs, load_summary
from analysis.metrics import lateral_error_mm_per_m, summarize_runs
from analysis.notebook import show_full_data
from analysis.plots import plot_heading, plot_power, plot_velocity, set_plot_style
from analysis.validation import validate_data

set_plot_style()
test_id = 'DT02'
```

```python
data = load_runs(root / 'data' / 'raw' / test_id)
if data.empty:
    display(Markdown('**Chưa có dữ liệu DT02.**'))
else:
    display(validate_data(data))
    display(summarize_runs(data))
    for run_id, run in data.groupby('run_id', sort=False):
        plot_velocity(run, f'{run_id} — vận tốc'); plt.show()
        plot_heading(run, f'{run_id} — heading'); plt.show()
        plot_power(run, f'{run_id} — điện'); plt.show()
    show_full_data(data)
```

```python
summary = load_summary(root / 'data' / 'run_summary.csv')
metrics = summary[summary.scenario_id.eq(test_id)].copy() if 'scenario_id' in summary else summary
display(metrics)
```
